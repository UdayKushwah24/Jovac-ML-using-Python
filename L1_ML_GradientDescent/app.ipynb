```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(x):
    return 3*x**2-4*x+5
```

```python
x = np.linspace(-2.5, 4 , 100)
y = fn(x)
```

```python
plt.plot(x, y)
plt.grid()
```

```python
def derivativeFn(x):
    return 6*x-4
```

```python
alpha = 0.1
xm = -2
```

```python
xm = xm - alpha*derivativeFn(xm)
print(xm)
```

```python
xml = [xm]
for i in range(20):
    xm = xm - alpha*derivativeFn(xm)
    xml.append(xm)
print(xml)
```

```python
plt.plot(x,y)

plt.plot(xml,fn(np.array(xml)), 'r' )
plt.grid()
```
