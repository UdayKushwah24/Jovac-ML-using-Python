```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y1 = [10, 5, 15, 20, 13, 16, 27, 18, 23, 30]
y2 = 2 * x + 3

plt.figure(figsize=(10, 6))
plt.plot(x, y1, marker='o', linestyle='--', color='#FF5733', linewidth=2.5, markersize=8, label='Price ')
plt.plot(x, y2, marker='D', linestyle='-', color='#3366FF', linewidth=2.5, markersize=7, label='F(x) = 2x + 3')

plt.xlabel('X Axis', fontsize=12, color='darkgreen')
plt.ylabel('Y Axis / f(X)', fontsize=12, color='darkblue')
plt.title('Fancy Line Plot - Price vs Function ', fontsize=16, color='purple')

plt.legend(loc='lower right', fontsize=11)
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.style.use('seaborn-v0_8-poster')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
sizes = [60, 15, 15, 10]
labels = ['CSE', 'Electronics', 'Mechanical', 'BioTech']
colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9']  # Stylish hex colors
explode = (0.1, 0.1, 0.1, 0.1)  # Pull all slices out slightly

# Create pie chart
fig, ax = plt.subplots(figsize=(8, 8), facecolor='#f0f0f0')  # Light background
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    labeldistance=1.15,
    colors=colors,
    autopct='%1.2f%%',
    shadow=True,
    explode=explode,
    startangle=140,
    textprops={'color': 'black', 'fontsize': 12, 'weight': 'bold'},
    wedgeprops={'linewidth': 2, 'edgecolor': 'white'}
)

# Title
plt.title('No. of Students vs Department',
          fontsize=16,
          weight='bold',
          color='#333333',
          pad=20)

# Layout
plt.tight_layout()
plt.show()
```
